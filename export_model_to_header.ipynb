Import stuff

```python
import numpy as np
import pickle
import pandas as pd

from keras.models import load_model
```

Load the model that was saved from the train_model script

```python
#load the model
model = load_model('model.h5')
weights = model.get_weights()

#see the diagram in the readme for descriptions on what W0, W1, W2, and W3 are
W0 = weights[0]
W1 = weights[1]
W2 = weights[2]
W3 = weights[3]
```

Load the scaling values and the threshold defined in the train_model script

```python
#get the number of features
num_feats = np.shape(W0)[0]
num_hidden_neurons = np.shape(W0)[1]

#load the feature scaler
scaler = pickle.load(open('standard_scaler.pickle', 'rb'))
mu = scaler.mean_
sigma = np.sqrt(scaler.var_)

#load the prob threshold for classification
thresh = pickle.load(open('thresh.pickle', 'rb'))
```

Write everything to a header file!

```python
with open('model_params.h', mode='wt', encoding='utf-8') as f: 
    f.write('//this file was generated by export_model_to_header.ipynb\n\n')

    f.write('#define NUM_FEATS ' + str(num_feats) + '\n')
    f.write('#define NUM_HIDDEN_NEURONS ' + str(num_hidden_neurons) + '\n\n')

    f.write('float W0[NUM_FEATS][NUM_HIDDEN_NEURONS] = \n{\n')
    for row in W0:
        f.write('\t{' + str(list(row))[1:-1] + '},\n')
    f.write('};\n\n')

    f.write('float W1[NUM_HIDDEN_NEURONS] = ')
    f.write('{' + str(list(W1))[1:-1] + '};')
    f.write('\n\n')

    f.write('float W2[NUM_HIDDEN_NEURONS] = {')
    for i,row in enumerate(W2):
        f.write(str(list(row))[1:-1])
        if i is not num_hidden_neurons-1:
            f.write(', ')
    f.write('};\n\n')

    f.write('float W3 = ' + str(list(W3))[1:-1] + ';')
    f.write('\n\n')

    f.write('float mu[NUM_FEATS] = ')
    f.write('{' + str(list(mu))[1:-1] + '};')
    f.write('\n\n')

    f.write('float sigma[NUM_FEATS] = ')
    f.write('{' + str(list(sigma))[1:-1] + '};')
    f.write('\n\n')

    f.write('float thresh = ' + str(thresh) + ';')
```
